## Goal
Build a simple loop where the agent can 'remember' context and 'use' a Python function.

```python
memory = []

def add_to_memory(role, text):
    memory.append(f"{role}: {text}")

def get_memory():
    return "\n".join(memory)
```

### Defining a Tool
Here is a simple calculator tool.

```python
def calculator(expression):
    try:
        return eval(expression)
    except:
        return "Error"
```

### The Agent Loop
Simulating a conversation where the agent decides to use a tool.

```python
user_input = "What is 25 * 48?"
add_to_memory("user", user_input)

# Simple keyword detection (In real life, LLM decides this)
if "*" in user_input or "calculate" in user_input:
    # Extract math (simplified)
    math_expr = user_input.replace("What is", "").replace("?", "").strip()
    result = calculator(math_expr)
    response = f"Calculated result: {result}"
else:
    response = "I can only calculate numbers right now."

add_to_memory("agent", response)
print(get_memory())
```
